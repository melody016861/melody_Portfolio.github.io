##第四題 :
> 畫出一個 $\frac{1}{x}$ 的圖形：使用 np.linspace 來定義 x 軸的範圍，並使用除法運算符 / 計算 y 軸的值，最後使用 plt.plot 來畫出函數圖形。注意需要特判 $x=0$ 的情況，因為無限大不能被計算。


$$\frac{1}{x}$$

```python
x = np.linspace(-10, 10, 200)
#從-10到10，有200個點在線性函數上。
```

```python
y = np.where(x != 0, 1 / x, np.inf)
```

```python
y = 1/x
```

```python
plt.plot(x, y, label=r"$\frac{1}{x}$")
plt.axhline(y=0, color='rosybrown', linestyle='--')  # 繪製 x 軸
plt.axvline(x=0, color='rosybrown', linestyle='--')  # 繪製 y 軸
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\frac{1}{x}$")
plt.grid(True)
```

##第六題 :


> 其實函數還可以更進一步「美化」, 比如說加粗線條、更換顏色等等。是否找得出要這樣做的方法?




```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(x) * np.cos(2*x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, color="purple", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("A Beautiful Function")
plt.grid(True)
```
