```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('grades.csv')

# Check the data types of each column
print(df.dtypes)

# Identify and handle non-numeric values
# For example, if the first column '姓名' contains non-numeric values, you can drop it
df = df.drop(columns=['姓名'])

# Convert all columns to numeric (excluding any non-numeric columns you want to keep)
df = df.apply(pd.to_numeric, errors='coerce')

# Now, calculate the correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
# 相關係數
df.corr()
```

```python
# 國文和數學的相關係數
df.國文.corr(df.數學)
```

```python
# dataframe名稱["新增欄位名稱"] = df[要總和的欄位的list].sum[1]
df["總級分"] = df[["國文", "英文", "數學", "社會", "自然"]].sum(1)
```

```python
df.head()
```

```python
# dataframe名稱["新增欄位名稱"] = dataframe名稱.項目.計算式
df["主科"] = df.數學*1.5 + df.英文
```

```python
df.head()
```

```python
# 依項目進行排序
# 由大到小 : dataframe名稱.sort_values(by="要排序項目", ascending=False)
# 由小到大 : dataframe名稱.sort_values(by="要排序項目")
df.sort_values(by="總級分", ascending=False).head(20) # 由大至小，前20個資料
```

```python
# 先用主科排序，如果主科相同，再用總級分排序
df.sort_values(by=["主科", "總級分"], ascending=False).head(20)
```

## 第二題

創建一個模擬的股票數據集，並使用 pandas 對股票價格進行分析。

```python
import pandas as pd
import numpy as np

# 創建模擬的股票數據集
np.random.seed(0)
dates = pd.date_range(start='2024-01-01', end='2024-04-01')
stock_prices = pd.DataFrame(np.random.normal(100, 10, (len(dates), 3)), columns=['Stock1', 'Stock2', 'Stock3'], index=dates)

# 計算每支股票的收益率
stock_returns = stock_prices.pct_change()

# 求最高價格和最低價格
max_prices = stock_prices.max()
min_prices = stock_prices.min()

print(stock_returns.head())
print(max_prices)
print(min_prices)
```
