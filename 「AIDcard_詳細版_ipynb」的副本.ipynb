## 3️⃣ 回收資訊和教育宣導

### 介紹

隨著環境問題日益嚴重，回收已成為我們生活中不可或缺的一部分。通過正確的回收，我們不僅可以減少垃圾焚化場的壓力，還可以保護自然資源。

隨著科技的進步和環境保護意識的提升，智慧回收裝置成為推動循環經濟和資源再利用的重要工具。智慧回收裝置不僅提高了回收的效率和準確性，還具有教育意義，能夠有效地提升公眾的環保意識和參與度。


希望看到這篇文章的你，可以更了解回收的重要性喔!

### 回收的重要性

1. **減少垃圾焚化場壓力**：
    - 垃圾焚化場是處理廢物的主要方式，但它們佔用了大量的土地資源。
    - 隨著垃圾量的增加，焚化場的壓力也在不斷增大，目前已到了供不應求的地步。

2. **保護自然資源**：
    - 回收有助於保護有限的自然資源，如金屬、木材和石油。
    - 透過回收，我們可以重新利用這些資源，減少對原材料的需求。

3. **減少污染**：
    - 垃圾中的有害物質會污染土壤和水源，對生態環境造成嚴重影響。
    - 正確回收有害物質可以避免它們對環境的污染。

### 如何進行有效的回收

1. **分類回收**：
    - 將垃圾分為可回收和不可回收兩大類。
    - 常見的可回收物包括紙張、塑料、金屬和玻璃。
    - 藉由智慧回收裝置的影像分類，我們可以更有效率的進行回收分類，回收正確率也會大幅提升。

2. **清潔回收物**：
    - 在丟棄可回收物之前，應確保它們已經被清潔乾淨。
    - 這樣可以提高回收物的質量，便於後續的加工處理。

3. **遵守當地回收規定**：
    - 每個地區的回收規定可能有所不同，應該了解並遵守當地的回收政策。
    - 這樣可以確保回收工作的有效性和規範性。

### 教育宣導的策略

1. **學校教育**：
    - 在學校課程中加入環境保護和回收的內容。
    - 通過實踐活動，如校內回收比賽，提高學生的環保意識。

2. **社區宣導**：
    - 組織社區講座和工作坊，向居民傳授回收知識。
    - 利用社交媒體和社區公告板發布回收資訊和提示。

3. **企業參與**：
    - 鼓勵企業參與回收計劃，並提供相應的獎勵措施。
    - 在辦公室內設置回收站，推動員工參與回收活動。

### 結論

回收不僅是解決垃圾問題的一種方式，更是保護環境、節約資源的重要手段。通過教育宣導，我們可以提高公眾的環保意識，促使更多人參與回收行動。讓我們共同努力，為保護地球環境貢獻一份力量吧!

````python
# @title 爬蟲後將圖片下載到本機中
import requests
from bs4 import BeautifulSoup
from PIL import Image, ImageDraw, ImageFont
from io import BytesIO
import os

def fetch_image_urls(query, max_links_to_fetch, headers):
    search_url = f"https://www.google.com/search?q={query}&tbm=isch"
    response = requests.get(search_url, headers=headers)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    image_elements = soup.find_all('img', limit=max_links_to_fetch)
    image_urls = []
    for img in image_elements:
        url = img.get('src') or img.get('data-src')
        if url and url.startswith('http'):
            image_urls.append(url)
    return image_urls

def download_and_label_image(url, label, headers):
    response = requests.get(url, headers=headers)
    img = Image.open(BytesIO(response.content)).convert('RGB')
    draw = ImageDraw.Draw(img)

    # 使用 textbbox 计算文本尺寸
    try:
        text_bbox = draw.textbbox((0, 0), label)
        position = (img.width - (text_bbox[2] - text_bbox[0]) - 10, img.height - (text_bbox[3] - text_bbox[1]) - 10)
        draw.text(position, label, (255, 255, 255))
    except Exception as e:
        print(f"Failed to label image: {e}")

    return img

def main():
    query = "paper recycling"
    max_images = 5
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
    save_folder = "labeled_images"
    os.makedirs(save_folder, exist_ok=True)

    print("Fetching image URLs...")
    image_urls = fetch_image_urls(query, max_images, headers)
    print(f"Found {len(image_urls)} images. Downloading and labeling...")

    for i, url in enumerate(image_urls):
        try:
            img = download_and_label_image(url, u"纸類", headers)
            img.save(os.path.join(save_folder, f"paper_recycling_{i + 1}.jpg"))
            print(f"Saved image {i + 1}")
        except Exception as e:
            print(f"Failed to process image {i + 1}: {e}")

    print("Done!")

if __name__ == "__main__":
    main()
````

````python
# @title 爬蟲回收分類_100images
query = "玻璃" # @param ["\u73BB\u7483", "\u7D19\u5F35", "\u7D19\u677F", "\u5851\u81A0", "\u91D1\u5C6C", "\u7D19\u5BB9\u5668", "\u4E00\u822C\u5783\u573E"] {allow-input: true}
import requests
from bs4 import BeautifulSoup
from PIL import Image, ImageDraw, ImageFont
from io import BytesIO
import os
from IPython.display import display

def fetch_image_urls(query, max_links_to_fetch, headers):
    search_url = f"https://www.google.com/search?q={query}&tbm=isch"
    response = requests.get(search_url, headers=headers)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    image_elements = soup.find_all('img', limit=max_links_to_fetch)
    image_urls = []
    for img in image_elements:
        url = img.get('src') or img.get('data-src')
        if url and url.startswith('http'):
            image_urls.append(url)
    return image_urls

def label_image(img, label):
    draw = ImageDraw.Draw(img)

    # 使用默認字體
    font = ImageFont.load_default()

    # 使用 textbbox 計算文本尺寸
    text_bbox = draw.textbbox((0, 0), label, font=font)
    position = (img.width - text_bbox[2] - 10, img.height - text_bbox[3] - 10)
    draw.text(position, label, (255, 255, 255), font=font)

def display_images_with_labels(image_urls, label):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

    for i, url in enumerate(image_urls):
        try:
            response = requests.get(url, headers=headers)
            img = Image.open(BytesIO(response.content)).convert('RGB')
            label_image(img, label.encode('utf-8'))  # 將標籤字符串編碼為UTF-8
            display(img)
            print(f"Displayed image {i + 1}")
        except Exception as e:
            print(f"Failed to process image {i + 1}: {e}")

def main():
    max_images = 100
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

    print("Fetching image URLs...")
    image_urls = fetch_image_urls(query, max_images, headers)
    print(f"Found {len(image_urls)} images. Displaying images with labels...")

    display_images_with_labels(image_urls, query)

if __name__ == "__main__":
    main()
````

````python
import PIL.Image
import pyheif

# 打开 HEIC 文件
heif_file = pyheif.read('metal.png')

# 将 HEIC 图像转换为 PIL 图像对象
img = PIL.Image.frombytes(
    heif_file.mode,
    heif_file.size,
    heif_file.data,
    "raw",
    heif_file.mode,
    heif_file.stride,
)

# 展示图像
img.show()
````

````python
model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content(img)

to_markdown(response.text)
````
